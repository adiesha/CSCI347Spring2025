```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features: sepal length, sepal width, petal length, petal width
y = iris.target  # Target: 0 (setosa), 1 (versicolor), 2 (virginica)

# Extract sepal length (X1) and sepal width (X2)
sepal_length = X[:, 0]  # First column: sepal length
sepal_width = X[:, 1]   # Second column: sepal width

# Create a scatter plot
plt.figure(figsize=(8, 6))
markers = ['o', '^', 's']  # Circle, triangle, square for the three classes
labels = iris.target_names
colors = ['gray', 'black', 'black']  # Colors for the classes (adjust as needed

plt.scatter(sepal_length, sepal_width, c=y)

# Set axis labels and limits
plt.xlabel('$X_1$ (Sepal Length)')
plt.ylabel('$X_2$ (Sepal Width)')
plt.xlim(4.3, 7.8)
plt.ylim(2.0, 4.4)

# Add axis-parallel lines (splits)
# Vertical lines: h0 at X1 = 5.8, h4 at X1 = 4.8, h5 at X1 = 6.3
plt.axvline(x=5.8, color='black', linestyle='--', label='$h_0$')
plt.axvline(x=4.8, color='black', linestyle='--', label='$h_4$')
plt.axvline(x=6.3, color='black', linestyle='--', label='$h_5$')

# Horizontal lines: h2 at X2 = 3.0, h3 at X2 = 3.5
plt.axhline(y=3.0, color='black', linestyle='--', label='$h_2$')
plt.axhline(y=3.5, color='black', linestyle='--', label='$h_3$')

# Add region labels (approximate positions based on the image)
plt.text(4.5, 4.2, '$R_1$', fontsize=12)
plt.text(5.0, 2.2, '$R_3$', fontsize=12)
plt.text(5.0, 2.8, '$R_4$', fontsize=12)
plt.text(6.0, 2.8, '$R_2$', fontsize=12)
plt.text(6.0, 3.8, '$R_5$', fontsize=12)
plt.text(7.0, 3.8, '$R_6$', fontsize=12)

# Add a marker for z (top-right corner)
plt.scatter([7.8], [4.4], marker='s', color='black', label='$z$')

# Add legend
plt.legend()

# Title
plt.title('(a) Recursive Splits')

# Show the plot
plt.grid(True)
plt.show()
```
