First load the dataset in the facebook_combined.txt file as a graph.

You can find this dataset under d2l Week 5 folder.

```python
import networkx as nx
G = nx.read_edgelist('facebook_combined.txt', delimiter=" ")
```

Determine how many nodes and edges are in this graph.

```python
print(G.number_of_nodes())
print(G.number_of_edges())
```

Draw the graph using networkx to get an idea about the dataset.

```python
nx.draw(G, node_size = 1, width=0.15)
```

If it takes too long to draw this, then take a sample of the graph and draw it.

```python
import random
rand_node_sample = random.sample(list(G.nodes), 800)
sub_graph = G.subgraph(rand_node_sample)
nx.draw(sub_graph, node_size = 0.5, width=0.15)
```

Plot the histogram of the degrees in the given graph.

```python
import matplotlib.pyplot as plt
plt.hist(dict(nx.degree(G)).values())
plt.xlabel("Degree")
plt.ylabel("Number of nodes")
```

What is the highest and lowest degrees available in the graph?

```python
print(f"Max degree is {max(dict(nx.degree(G)).values())} and min degree is {min(dict(nx.degree(G)).values())}")
```

Compute the degree centrality of the graph.

Go throught the networkx library and find if they have a method for this, and check if you can get the answer that you are looking for. Hint: Their implementation may be slightly different.

```python
print(sum(nx.degree_centrality(G).values())/G.number_of_nodes())
```
